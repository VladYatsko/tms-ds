### Task 1. Matrix calculations with numpy

##### Option 1.

```python
import numpy as np


MATRIX_A = np.array([[1, 2, 4], [7, 5, -2], [-2, 1, 3]])
MATRIX_B = np.array([[-1, 0, 3], [2, -6, 1], [-7, 1, 5]])
MATRIX_C = np.array([[-2, 3], [4, 0], [7, -1]])
```

##### A + B

```python
MATRIX_D = MATRIX_A + MATRIX_B

MATRIX_D
```

##### A - B

```python
MATRIX_D = MATRIX_A - MATRIX_B

MATRIX_D
```

##### A * C

```python
MATRIX_D = MATRIX_A.dot(MATRIX_C)

MATRIX_D
```

##### A * B * C

```python
MATRIX_D = MATRIX_A.dot(MATRIX_B) @ MATRIX_C

MATRIX_D
```

##### Uploading the image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('dog.jpg')
img_cleansed = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(img_cleansed)
```

##### Initializing kernel

```python
kernel_size = (3, 3)
sigma = 1

kernel = cv2.getGaussianKernel(kernel_size[0], sigma)
kernel = np.outer(kernel, kernel)
```

##### Applying algorythm

```python
blurred = cv2.filter2D(img_cleansed, -1, kernel)
```

##### Saving the image

```python
cv2.imwrite('blurred.jpg', blurred)
cv2.destroyAllWindows()
```

```python
plt.imshow(blurred)

img_cleansed = cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(img_cleansed)
```
